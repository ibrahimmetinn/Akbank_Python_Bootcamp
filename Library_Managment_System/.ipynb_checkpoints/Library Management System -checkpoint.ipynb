```python
class Library:
    def __init__(self):
        '''
        #books.txt adında dosya oluşturuyororuz ve bunu her değişkene özel kılmak 
        adına __init__fonksiyonu içerisinde yapıyoruz.
        Aynı zamanda bunu her seferinde tek tek yazmamak adına dosya (file adında tanımlıyoruz)
        '''
        self.file_path = "books.txt"
        self.file = open(self.file_path, "a+")  
        
        '''
        Dosyayı a+ modunda açıyoruz dosya modları içerisinde, 6 adet mod mod vardır .
        w(write),r(read),a(append),x(create),t(text),b(binary)
        neden x de değilde a+ modunda açtığımızı düşünüyor olabilirsiniz. 
        x metin dsoyası oluşturur fakat tekrar çalıştırdığımızda
        hata meydana gelir. Ve veri ekleme ypamayız7
        '''
           
    def __del__(self):
        '''
        Bellekte sorunlar olmaması için close fonksiyonunu kullanmak zorundayız.
        Her yazmış olduğumuz veri sonrasında dosya kapatma fonksiyonu.
        
        '''
        self.file.close()
    def list_books(self):
        print("================================================")
        '''
        seek() fonksiyonu parametre ile beraber kullanılır. Bu noktada parametre file(dosya)dır. 
        seek fonksiyonu sayesinde txt dosyası içerisinde istediğimiz konuma gidebilme olanağı sağlıyor.
        Burada 0 kullanmamızın nedeni dosyada 0 yani başlangıç noktasına gidebilmektir.(içerisindeki rakam byte temsil eder.) 
        '''
        self.file.seek(0)
        '''
        read fonksiyonu bir file handling konusudur. Yani oluşturduğumuz metin dosyasını okur. 
        Üç tip okuma methodu vardır. read,readline,readlines read tüm dosyayı okur. readline satır satır okur.
        readlines verileri liste tipinde okur. Bu noktada tüm dosyayı okumak daha anlamlı olacaktır.
        
        splitlines methodu okuduğumuz metni listeye böler. Burada kritik nokta şu splitlines nereden bölüyor. 
        book_1 = f"{title},{author},{release_date},{pages}\n"
        bu f-string methodu ile yazılmış olan kodun sonunda yani satır atlama noktasından itibaren bölüyor ki,
        kayıt altına alınmış her bir kitabı tek bir listede tutmamıza olanak sağlıyor.
        ----------------------------------------------------------------------------------
        book1=kavgam,adolf hitler,1925,596------------------------->\n(böl ve liste oluştur [kavgam,adolf,hitler,1925,596])
              silmarillion,tolkien,1977,682------------------------->\n(böl ve liste oluştur[silmarillion,tolkien,1977,682])
        
        Tek bir liste içerisinde tutmamızın sebebi remove fonksiyonunu kullanırken sadece kitap ismini yazdığımız zaman 
        o kitap ile ilgili tüm verilerin ortadan kaldırılmasını istiyoruz.
        ----------------------------------------------------------------------------------
        '''
        books = self.file.read().splitlines()
        for book in books: 
            '''
            # Şimdi bölmüş olduğumuz liste içerisinde her bir eleman üzerinde dönen bir book değişkeni olsun. 
            book değişkeni ',' ile bölüp her bir değişkeni ayrı ayrı bir değişkene atıyoruz ki. PDF dosyasında istenilen
            Book: (title), Author:(Author) şeklinde yazabilelim. Ve kullanıcı sorgulama yaptığında bu şekilde bir görüntüleme sağlanır.
            '''
            book_info = book.split(',')
            title, author, release_date, pages = book_info
            print(f"Title: {title}, Author: {author}")
            print("================================================")
            
            
        

    def add_book(self):
        
        '''
        Kitapları eklediğimiz fonksiyon.
        
        '''
        title = input("Enter the book title (Kitap ismini giriniz)                :")
        print("-----------------------------------------------------------------")
        author = input("Enter the author's name (Kitap yazarını giriniz)          :")
        print("-----------------------------------------------------------------")
        release_date = input("Enter the release date (Yayımlanma tarihini giriniz):")
        print("-----------------------------------------------------------------")
        pages = input("Enter the number of pages (Sayfa sayısını giriniz)         :")
        print("-----------------------------------------------------------------")

        book_1 = f"{title},{author},{release_date},{pages}\n"
        
        self.file.write(book_1)
        print(f"Book '{title}' ADDED")

    def remove_book(self):
        title_to_remove = input("Enter the title of the book to remove: ")
        '''
        Kitapları remove ettiğimiz bölüm.
        Tekrar imleci 0 konumuna getirdik.
        Daha sonra tekrar list fonksiyonunda olduğu gibi  line line kaldıracağımız kitap ismini arayıp,
        splitlines() ile böldüğümüz liste içerisinde for döngüsü ile her bir elemanı sorgulayıp title_remove
        değişkenini yakalamaya çalışıyoruz. 
        içerisinde arama yaptırım eşleşen veriyi siliyoruz.
        '''

        self.file.seek(0)
        
    
        books = self.file.read().splitlines()
        
        for book in books:
            if title_to_remove in book:
                books.remove(book)
                break

        self.file.seek(0)
        self.file.truncate()
        self.file.writelines("\n".join(books))

        print(f"Book '{title_to_remove}' removed successfully.")
        print("________________________________________________")


lib = Library()

# Menu
while True:
    print("\n************************ MENU **************************\n")
    
    print("|   1) List Books","2) Add Book","3) Remove Book","q) Exit   |")
    print("_______________________________________________________")
    
    Enter = input("Enter_your choice (q) exit): ")

    if Enter == "1":
        lib.list_books()
    elif Enter == "2":
        lib.add_book()
    elif Enter == "3":
        lib.remove_book()
    elif Enter == "q":
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 3. If you want exit the program  press enter q  ")
```
